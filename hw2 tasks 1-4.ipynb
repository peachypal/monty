```python
def proper_title_case(s):
    nocaps = ["the", "in", "on", "a", "an",'of'] # This needs to be extended.
    #Your code goes below this comment
    ourstring=s.split()
    returnstring=""
    for i in range(len(ourstring)):
        for k in range(len(nocaps)):            
            ourstring[i]=ourstring[i].lower()                                
            if ourstring[i]!=nocaps[k] or i==0:
                x=ourstring[i]
                x=x.capitalize()
                ourstring[i]=x
            else:
                break
        returnstring+=ourstring[i]+" "
    return returnstring#here goes what you want the function to return
print(proper_title_case('An ant on a table'))#to test your function
print(proper_title_case('The Cat In the Hat'))#to test your function
print(proper_title_case('Property Of Michael Scott Paper Company'))
```

```python
def mean(vals):
    """Return the mean of the values in vals, presupposed to be 
    numeric (float, int, or long)."""
    return sum(vals)/len(vals)#here goes what you want the function to return

print(mean([1,5,1.1]))#to test your function
```

```python
import math

def sd(vals):
    """Return the standard deviation of the values in vals, 
    presupposed to be numeric (float, int)."""
    #Your code goes below this comment, use the mean fuction defined in the previous task
    mean=sum(vals)/len(vals)
    y=0.0
    for i in range(len(vals)):
       y+=vals[i]*vals[i]-2*vals[i]*mean+mean*mean
    y=y/len(vals)
    y=math.sqrt(y)
    return y#here goes what you want the function to return

print(sd([0,1,2]))#to test your function
```
